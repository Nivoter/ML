
Задание

Отобразите график ящика с усами с помощью функции sns.boxplot(), отображая один из показателей HP, Attack или Defense от поколения и легендарности.

```
# TODO - отобразите график ящика с усами
fig, ax = plt.subplots(nrows=1, ncols=3, squeeze=False, figsize=[15, 10])
sns.boxplot(x = df['Generation'],y = df['HP'],hue = df['Legendary'],ax=ax[0,0])
ax[0,0].grid(True)
sns.boxplot(x = df['Generation'],y = df['Attack'],hue = df['Legendary'],ax=ax[0,1])
ax[0,1].grid(True)
sns.boxplot(x = df['Generation'],y = df['Defense'],hue = df['Legendary'],ax=ax[0,2])
ax[0,2].grid(True)
```
